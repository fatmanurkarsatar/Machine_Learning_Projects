```python
#kütüphaneleri import ettik.
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#DataFrame'i yalnızca 'temp_min' sütununun minimum değere eşit olduğu satırları içerecek şekilde filtreler.
df[df['temp_min']==min(df.temp_min)]
```

```python
#DataFrame'i yalnızca 'temp_max' sütununun maximum değere eşit olduğu satırları içerecek şekilde filtreler.
df[df['temp_max']==max(df.temp_max)]
```

```python
#kütüphaneleri import ediyoruz
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.preprocessing import LabelEncoder

df.info()
```

```python
#bu kod verisetimizdeki hava durumu sütununun orjinal kategorik etiketlere karşılık gelen sayısal gösterimlerle(benzersiz) değiştirilecek.
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['weather']=le.fit_transform(df['weather'])
```

```python
#eğitim ve tahmin yapıyoruz
x = df[['temp_min', 'temp_max', 'precipitation', 'wind']]
y=df['weather']
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

nb_model = GaussianNB()
nb_model.fit(X_train, y_train)

y_pred = nb_model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
classification_rep = classification_report(y_test, y_pred, zero_division=1)


print(f"Accuracy: {accuracy:.2f}")
```
